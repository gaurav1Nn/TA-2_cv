```python
# Cell 1: Import libraries and set up inline plotting
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cell 2: Load and process the image
image = cv2.imread('building.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Harris Corner Detection
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Dilate to mark corners
dst = cv2.dilate(dst, None)

# Threshold to find corner locations
threshold = 0.01 * dst.max()
corner_coords = np.argwhere(dst > threshold)

# Draw larger dots (circles) at corner locations
for y, x in corner_coords:
    cv2.circle(image, (x, y), radius=5, color=(0, 255, 255), thickness=-1)  # radius=5, filled circle

# Display the result
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection (Larger Dots)')
plt.axis('off')
plt.show()
```
